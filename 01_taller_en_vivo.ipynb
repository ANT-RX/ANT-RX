# Red Neuronal (Feed Forward Neural Network)


```python
class RedNeuronal:
  def __init__(self, entrada=8, capas=2, neuronas_por_capa=4, salida=4, seed=None):
    self.entrada = entrada
    self.capas = capas
    self.neuronas_por_capa = neuronas_por_capa
    self.salida = salida
    self.rng = np.random.default_rng(seed)
    self.pesos, self.biases = self.crear_parametros()


  def crear_parametros(self):
    pesos = []
    biases = []
    tamanos = [self.entrada] + [self.neuronas_por_capa] * self.capas + [self.salida]

    for indice in range(len(tamanos) - 1):
      entradas_capa = tamanos[indice]
      salidas_capa = tamanos[indice + 1]
      pesos.append(self.rng.uniform(-1, 1, size=(entradas_capa, salidas_capa)))
      biases.append(self.rng.uniform(-1, 1, size=(salidas_capa,)))

    return pesos, biases


  def entrenar(self):
    self.crear_pob


  def forward(self, entradas):
    activaciones = np.array(entradas, dtype=float)

    for indice_capa, (pesos_capa, biases_capa) in enumerate(zip(self.pesos, self.biases)):
      z = activaciones @ pesos_capa + biases_capa
      if indice_capa == len(self.pesos) - 1:
        activaciones = z
      else:
        activaciones = np.tanh(z)

    return activaciones


  def obtener_genoma(self, pesos, sesgos):
    partes = []
    for pesos_capa, biases_capa in zip(pesos, sesgos):
      partes.append(pesos_capa.flatten())
      partes.append(biases_capa.flatten())
    return np.concatenate(partes)


  def cargar_genoma(self, genoma):
    genoma = np.array(genoma, dtype=float)
    indice = 0

    for capa in range(len(self.pesos)):
      forma_pesos = self.pesos[capa].shape
      numero_pesos = self.pesos[capa].size
      self.pesos[capa] = genoma[indice:indice + numero_pesos].reshape(forma_pesos)
      indice += numero_pesos

      forma_biases = self.biases[capa].shape
      numero_biases = self.biases[capa].size
      self.biases[capa] = genoma[indice:indice + numero_biases].reshape(forma_biases)
      indice += numero_biases

    if indice != len(genoma):
      raise ValueError("El tamano del genoma no coincide con la arquitectura de la red")
```

# Algoritmo Genético (para encontrar los parámetros óptimos de la red neuronal)



```python
class Neuroevolucion:
  def __init__(self, tasa_mutacion, tamano_poblacion, num_generaciones, max_pasos, entornos_de_prueba) -> None:
    self.tasa_mutacion = tasa_mutacion
    self.max_pasos = max_pasos
    self.tamano_poblacion = tamano_poblacion
    self.num_generaciones = num_generaciones
    self.acciones = [(-1, 0), (1, 0), (0, 1), (0, -1)]
    self.direcciones = {(-1, 0):"ARRIBA", (1, 0):"ABAJO", (0, 1):"DERECHA", (0, -1):"IZQUIERDA"}
    self.entornos_de_prueba = entornos_de_prueba


  def mostrar_poblacion(self):
    for i in range(self.tamano_poblacion):
      print(self.poblacion[i])



  def seleccion(self, numero_candidatos):
    maximo_fitness = float("-inf")
    mejor_padre = None

    for i in range(numero_candidatos):
      indice = random.randint(0, self.tamano_poblacion - 1)
      padre = self.poblacion[indice]
      fitness = self.poblacion_fitness[indice]

      if fitness > maximo_fitness:
        maximo_fitness = fitness
        mejor_padre = padre

    return mejor_padre



  def fitness(self, pasos, min_distancia, gano):
    area = self.entorno.ancho * self.entorno.altura
    if gano:
      return area - pasos
    elif gano == False:
      return -area - min_distancia
    return -min_distancia


  def checar_derrota(self):
    if self.posicion_actual_agente_y < 0 or self.posicion_actual_agente_x < 0 or self.posicion_actual_agente_y >= self.entorno.altura or self.posicion_actual_agente_x >= self.entorno.ancho:
      return True
    if self.entorno.entorno[self.posicion_actual_agente_y][self.posicion_actual_agente_x] == "O":
      return True
    if (self.posicion_actual_agente_y, self.posicion_actual_agente_x) in self.visitados:
      return True
    return False


  def distancia_manhattan(self):
    return abs(self.entorno.posicion_meta_y - self.posicion_actual_agente_y) + abs(self.entorno.posicion_meta_x - self.posicion_actual_agente_x)


  def probar_solucion(self, genoma, entorno):
      self.red_neuronal.cargar_genoma(genoma)
      self.entorno = entorno
      self.posicion_actual_agente_x = entorno.posicion_inicial_agente_x
      self.posicion_actual_agente_y = entorno.posicion_inicial_agente_y
      self.visitados = {(self.posicion_actual_agente_y, self.posicion_actual_agente_x)}

      movimientos = []
      posiciones = [(self.posicion_actual_agente_y, self.posicion_actual_agente_x)]

      for paso in range(self.max_pasos):
        entradas = self.observacion() + [
            (self.entorno.posicion_meta_y - self.posicion_actual_agente_y) / self.entorno.altura,
            (self.entorno.posicion_meta_x - self.posicion_actual_agente_x) / self.entorno.ancho]
        salidas = self.red_neuronal.forward(entradas)
        accion = int(np.argmax(salidas))
        movimiento = self.salida_a_accion[accion]
        movimientos.append(self.direcciones[movimiento])

        self.mover()
        posiciones.append((self.posicion_actual_agente_y, self.posicion_actual_agente_x))

        if self.checar_derrota():
          return movimientos, posiciones, False

        self.visitados.add((self.posicion_actual_agente_y, self.posicion_actual_agente_x))

        if self.entorno.entorno[self.posicion_actual_agente_y][self.posicion_actual_agente_x] == "M":
          return movimientos, posiciones, True

      return movimientos, posiciones, False
```
